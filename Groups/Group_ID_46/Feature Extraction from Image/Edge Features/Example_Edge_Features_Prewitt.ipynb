# Feature Extraction from Image (Edge Detection)
---

### Package Name: Edge_Features
---

## Methods:

### 1. horizontal_edge_prewitt(image):
This method returns the horizontal edge features extracted from a grayscale image using prewitt_h().

#### Parameters:

- Image: It takes a grayscale image having dimensions M x N

### 2. vertical_edge_prewitt(image):
This method returns the vertical edge features extracted from a grayscale image using prewitt_v().

#### Parameters:

- **Image:** It takes a grayscale image having dimensions M x N
---

## Example:

```python
import cv2
import matplotlib.pyplot as plt
from Edge_Features import Edge_Features
```

```python
# Creating object by calling class Edge_Features
features = Edge_Features()

# Read input image as grayscale image
image = cv2.imread('flat_chessboard.png', cv2.IMREAD_GRAYSCALE)

# Passing image to horizontal_edge_prewitt() to get horizontal edge features
horizontal_edges = features.horizontal_edge_prewitt(image)

# Passing image to vertical_edge_prewitt() to get vartical edge features
vertical_edges = features.vertical_edge_prewitt(image)

all_edges = horizontal_edges + vertical_edges

#Display input image and extracted edge features
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12,4))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Chessboard Image")
axes[1].imshow(horizontal_edges, cmap='gray')
axes[1].set_title('Horizontal Edges')
axes[2].imshow(vertical_edges, cmap='gray')
axes[2].set_title('Vertical Edges')
axes[3].imshow(all_edges, cmap='gray')
axes[3].set_title('All Edges')
plt.tight_layout()
plt.show()
```
