# Feature Extraction from Image (Edge Detection)
---

### Package Name: Edge_Features
---

## Methods:

### edge_detection_canny(image, lower, upper):

This method returns the edge features, extracted from a colored image, in the form of a 2D matrix. It preprocesses input image and then uses 'Canny Edge Detection' to extract the edge features from the image.

#### Parameters:

- **Image:** It takes a 3 dimensional image having dimensions M x N x 3. Where, M = height of image; N = width of image; 3 = number of channels.

- **Lower:** Optional parameter that takes lower threshold value for the canny algorithm

- **Upper:** Optional parameter that takes upper threshold value for the canny algorithm
---

## Example:

```python
import cv2
import matplotlib.pyplot as plt
from Edge_Features import Edge_Features
```

```python
# Creating object by calling class Edge_Features
features = Edge_Features()

# Using cv2.imread('') to read input image
image_1 = cv2.imread('road_image.jpg')
image_2 = cv2.imread('rubiks_cube.png')

# Convert image from BGR format to RGB format
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
image_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB)

# Passing image to edge_detection() to get edge features
edges_1 = features.edge_detection_canny(image_1)
edges_2 = features.edge_detection_canny(image_2,20,85)

#Display input image and extracted edge features
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
axes[0,0].imshow(image_1)
axes[0,0].set_title('Road Image')
axes[1,0].imshow(image_2)
axes[1,0].set_title("Rubik's Cube Image")
axes[0,1].imshow(edges_1, cmap='gray')
axes[0,1].set_title('Road Edges')
axes[1,1].imshow(edges_2, cmap='gray')
axes[1,1].set_title("Rubik's Cube Edges")
plt.tight_layout()
plt.show()
```
